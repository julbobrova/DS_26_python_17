```python
# Задание 1
# Пункт 1. Пользователь по команде “p” может узнать владельца документа по его номеру +
# Пункт 2. Пользователь по команде “s” может по номеру документа узнать на какой полке он хранится +
# Пункт 3. Пользователь по команде “l” может увидеть полную информацию по всем документам +
# Пункт 4. Пользователь по команде “as” может добавить новую полку +
# Пункт 5. Пользователь по команде “ds” может удалить существующую полку из данных (только если она пустая)+
# Задание 2
# Пункт 1. Пользователь по команде “ad” может добавить новый документ в данные+
# Пункт 2. Пользователь по команде “d” может удалить документ из данных+
# Пункт 3. Пользователь по команде “m” может переместить документ с полки на полку


```

```python
# Исходные данные
documents = [
{'type': 'passport', 'number': '2207 876234', 'name': 'Василий Гупкин'},
{'type': 'invoice', 'number': '11-2', 'name': 'Геннадий Покемонов'},
{'type': 'insurance', 'number': '10006', 'name': 'Аристарх Павлов'}
]

directories = {
'1': ['2207 876234', '11-2'],
'2': ['10006'],
'3': []
}
```

```python
def p_comnd(number):
    """
    Поиск владельца по номеру документа
    """
    for DocDict in documents:
        if DocDict.get('number') == number:
            return  DocDict.get('name')
            break
```

```python
def s_comnd(number):
    """
    Поиск номера полки хранения по номеру документа
    """
    for shelf in directories:
        for numb_ in directories.get(shelf):
            if numb_ == number:
                return shelf
                break
```

```python
def l_comnd():
    """
    Информация по всем документам
    """
    print("Текущий перечень документов: ")
    for DocDict in documents:
            print(f" №: {DocDict.get('number')}, тип: {DocDict.get('type')}, владелец: {DocDict.get('name')}, полка хранения: {s_comnd(DocDict.get('number'))}")
   
```

```python
def as_comnd(shelf_num):
    """
    Добавление новой полки
    """
    if shelf_num in directories:
        print("Такая полка уже существует. Текущий перечень полок:", end=' ')
    else:
        print("Полка добавлена. Текущий перечень полок:", end=' ')
    directories.setdefault(shelf_num) 
    s = str(list(directories.keys()))
    print(s[1:len(s)-1].replace("'",""))
```

```python
def ds_comnd(shelf_num):
    """
    Удаление полки, если она пустая!
    """
    if shelf_num not in directories:
        print("Такая полка не существует.", end=' ')
    elif directories.get(shelf_num) != None:
        print("На полке есть документа, удалите их перед удалением полки.", end=' ')
    else: 
        del directories[shelf_num]
        print("Полка удалена.", end=' ')
    s = str(list(directories.keys()))
    print("Текущий перечень полок: ", s[1:len(s)-1].replace("'",""))
```

```python
def ad_comnd(*new_doc):
    """
    Добавление нового документа. Полка для добавления должна существовать!
    Первый этап - проверяем наличие полки. Если нет - выход
    Второй этап - добавляем документ в список документов
    """
    if new_doc[3] not in directories:
        print("Такая полка не существует. Для добавлении полки используйте команду as")
    else:
        DocDict['type'] = new_doc[0]
        DocDict['number'] = new_doc[1]
        DocDict['name'] = new_doc[2]
        documents.append(DocDict)
        directories[new_doc[3]].append(new_doc[1])
        print("Документ добавлен.")
    
```

```python
def d_comnd(number):
    """
    Удаление документа
    """
    for DocDict in documents:
        if DocDict.get('number') == number:
            documents.remove(DocDict)
            s_comnd(number)
            directories[s_comnd(number)].remove(number) 
            break
    print("Документ не найден в базе")    
```

```python
def m_comnd(number, shelf):
    """
    Переместить документ с номером number  на новую полку  shelf
    """
    if p_comnd(number) == 0:
        result = 0
    elif shelf not in directories:
        result = 1
    else:
        directories[s_comnd(number)].remove(number) 
        directories[shelf].append(number)
        result = 2
    return result    
```

```python
print("Список доступных команд для помощника")
print(" ad - добавить новый документ в данные")
print(" as - может добавить новую полку")
print("  d - удалить документ из данных")
print(" ds - может удалить существующую полку из данных (только если она пустая)")
print("  l - увидеть полную информацию по всем документам")
print("  p - по номеру документа узнать его владельца")
print("  s - по номеру документа узнать на какой полке он хранится")
print("  m - переместить документ с полки на полку ")
print("  q - выйти из программы помощника \n\n")
cmd =str(input("Введите команду: \n"))
while cmd !="q":
    if cmd == "p":
        number = str(input("Введите номер документа: \n"))
        if p_comnd(number) == None:
            print("Документ не найден в базе")
        else:
            print(f"Владелец документа: {p_comnd(number)}")
    elif cmd == "s":
        number = str(input("Введите номер документа: \n"))
        if s_comnd(number) == None:
            print("Документ не найден в базе") 
        else:
            print(f"Документ хранится на полке: {s_comnd(number)}")
    elif cmd == "l":    
        l_comnd()
    elif cmd == "as":
        number = str(input("Введите номер полки: \n"))
        as_comnd(number)
    elif cmd == "ds":
        number = str(input("Введите номер полки: \n"))
        ds_comnd(number)
    elif cmd == "ad":
        number = str(input("Введите номер документа: \n"))
        type_ = str(input("Введите тип документа: \n"))
        owner = str(input("Введите владельца документа: \n"))
        shelf = str(input("Введите номер полки: \n"))       
        ad_comnd(number, type_, owner, shelf)
        l_comnd()    
    elif cmd == "d":
        number = str(input("Введите номер документа: \n"))
        d_comnd(number)
        l_comnd() 
    elif cmd == "m":    
        number = str(input("Введите номер документа: \n"))
        shelf = str(input("Введите номер новой полки: \n"))
        if m_comnd(number, shelf) == 1:
            print("Такой полки не существует. ", end=' ') 
            s = str(list(directories.keys()))
            print("Текущий перечень полок: ", s[1:len(s)-1].replace("'",""))
        else:
            if m_comnd(number, shelf) == 0:
                print("Документ не найден в базе")
            else:        
                print("Документ перемещен")
            l_comnd()
                
    cmd =str(input("Введите команду: \n"))
print("Работа помощника завершена")    
```
